```python
nrlz=50
yarr=np.logspace(-6,-4.8,10)
```

```python
start=time.time()
cl_char=collections.OrderedDict()
cl_char["snr"]=np.zeros((len(yarr),6,nrlz))
cl_char["yc"]=np.zeros((len(yarr),6,nrlz))
cl_char["thetac"]=np.zeros((len(yarr),6,nrlz))
for iy,yc in enumerate(yarr):
    data,noise,yc_true=cnc.simulate_ymap(12.,yc,xy=[123,32])  
    pure_ymap=data-noise
    for i in range(nrlz):
        data = pure_ymap + cnc.simulate_noise()
        soln_ak=cnc.mf_ideal_all_known(data,12.,xy=[123,32])
        soln_sk=cnc.mf_ideal_size_known(data,12.)
        soln_lk,res_lk=cnc.mf_ideal_loc_known(data,xy=[123,32])
        soln,res=cnc.mf_ideal(data)
        soln_r,res_r=cnc.mf_real(data)
        soln_ir,res_ir=cnc.mf_real_iterative(data)
        cl_char["snr"][iy,:,i]=np.array([soln_ak[0],soln_sk[0],soln_lk[0],soln[0],soln_r[0],soln_ir[0]])
        cl_char["yc"][iy,:,i]=np.array([soln_ak[1],soln_sk[1],soln_lk[1],soln[1],soln_r[1],soln_ir[1]])
        cl_char["thetac"][iy,:,i]=np.array([soln_ak[3],soln_sk[3],soln_lk[3],soln[3],soln_r[3],soln_ir[3]])
        
print time.time()-start
```

```python
char_stat=collections.OrderedDict()
char_stat["snr"]=np.zeros((len(yarr),6,2))
char_stat["yc"]=np.zeros((len(yarr),6,2))
char_stat["thetac"]=np.zeros((len(yarr),6,2))

for iy,yc in enumerate(yarr):
    for i in range(6):
        for k in cl_char.keys():
            char_stat[k][iy,i,0]=np.mean(cl_char[k][iy,i,:])
            char_stat[k][iy,i,1]=np.std(cl_char[k][iy,i,:])
        
```

```python
char="yc"
figure()
plot(char_stat["snr"][:,0,0],char_stat[char][:,1,0]/char_stat[char][:,0,0],label="Size known")
plot(char_stat["snr"][:,0,0],char_stat[char][:,2,0]/char_stat[char][:,0,0],label="Loc. known")
plot(char_stat["snr"][:,0,0],char_stat[char][:,3,0]/char_stat[char][:,0,0],label="No known")
plot(char_stat["snr"][:,0,0],char_stat[char][:,4,0]/char_stat[char][:,0,0],label="Real")
plot(char_stat["snr"][:,0,0],char_stat[char][:,5,0]/char_stat[char][:,0,0],"--",label="Real iter.")

semilogy()
legend(loc=0)
axhline(1,linestyle="dashed",c="k")


char="thetac"
figure()
plot(char_stat["snr"][:,0,0],char_stat[char][:,1,0]/char_stat[char][:,0,0],label="Size known")
plot(char_stat["snr"][:,0,0],char_stat[char][:,2,0]/char_stat[char][:,0,0],label="Loc. known")
plot(char_stat["snr"][:,0,0],char_stat[char][:,3,0]/char_stat[char][:,0,0],label="No known")
plot(char_stat["snr"][:,0,0],char_stat[char][:,4,0]/char_stat[char][:,0,0],label="Real")
plot(char_stat["snr"][:,0,0],char_stat[char][:,5,0]/char_stat[char][:,0,0],"--",label="Real iter.")
# semilogy()
legend(loc=0)
axhline(1,linestyle="dashed",c="k")
```
