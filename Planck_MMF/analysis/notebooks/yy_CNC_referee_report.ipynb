```python
planckyy=np.loadtxt(datapath + "/maps/planck2015_yy_ps.txt")
bmin=planckyy[:,3]
bmax=planckyy[:,4]

# Data from Boris's papers
planck_yy=np.loadtxt(datapath + "/maps/bolliet2018.txt")
```

```python
yy=loadtxt("../../results/yy_spec/snr6_lmax1500_mlmax4096_apow15_psmask/data/total_yy_spec.txt")
yy_rc=loadtxt("../../results/yy_spec/snr6_lmax1500_mlmax4096_apow15_psmask/data/resolved_yy_spec.txt")
yy_urc=loadtxt("../../results/yy_spec/snr6_lmax1500_mlmax4096_apow15_psmask/data/unresolved_yy_spec.txt")
```

```python
figure()
errorbar(yy[:,2],yy[:,3],yy[:,4],c="k",lw=2,label="Total",zorder=0)
plot(yy[:,2],yy_rc[:,3]+yy_urc[:,3],"--",label=r"$C_{\ell}^{\rm RC} + C_{\ell}^{\rm uRC}$",lw=2,c="r")
plot(yy[:,2],yy_urc[:,3],label=r"$C_{\ell}^{\rm uRC}$",lw=2,c="g")
plot(yy[:,2],yy_rc[:,3],label=r"$C_{\ell}^{\rm RC}$",lw=2,c="b")
ylabel(r"$10^{12}$" + r" $\ell(\ell+1)C_{\ell}/(2 \pi)$",fontsize=12)
xlabel("multipole, $\ell$",fontsize=12)
loglog()
grid(which="both",linestyle="dashed",alpha=0.6)
legend(loc=0)
figname=result_path + "compare_urc_rc_tot.pdf"
savefig(figname,bbox_inches="tight")
```

```python
rtot_err=np.sqrt((yy_rc[:,4]**2. + yy_urc[:,4]**2.)/yy[:,3]**2. + (yy[:,4]**2.)*((yy_rc[:,3]+yy_urc[:,3])**2)/(yy[:,3]**4.))
rrc_err=np.sqrt((yy_rc[:,4]**2.)/yy[:,3]**2. + (yy[:,4]**2.)*(yy_rc[:,3]**2)/(yy[:,3]**4.))
rurc_err=np.sqrt((yy_urc[:,4]**2.)/yy[:,3]**2. + (yy[:,4]**2.)*(yy_urc[:,3]**2)/(yy[:,3]**4.))
                

figure()
axhline(1,linestyle="--",c="k",lw=2)
errorbar(yy[:,2],rtot,yerr=rtot_err,lw=2,label=r"$(C_{\ell}^{\rm RC} + C_{\ell}^{\rm uRC})/C_{\ell}^{\rm Total}$",c="r")
errorbar(yy[:,2],rurc,yerr=rurc_err,lw=2,label=r"$C_{\ell}^{\rm uRC}/C_{\ell}^{\rm Total}$",c="g")
errorbar(yy[:,2],rrc,yerr=rrc_err,lw=2,label=r"$C_{\ell}^{\rm RC}/C_{\ell}^{\rm Total}$",c="b")

ylabel(r"Ratio of $yy$-power spectrum",fontsize=12)
xlabel("multipole, $\ell$",fontsize=12)
grid(which="both",linestyle="dashed",alpha=0.6)
legend(loc=0)
semilogx()

figname=result_path + "compare_ps_ratio_with_error.pdf"
savefig(figname,bbox_inches="tight")
```

```python
rtot=(yy_rc[:,3]+yy_urc[:,3])/yy[:,3]
rrc=(yy_rc[:,3])/yy[:,3]
rurc=(yy_urc[:,3])/yy[:,3]
figure()
axhline(1,linestyle="--",c="k",lw=2)
plot(yy[:,2],rtot,lw=2)
plot(yy[:,2],rrc,lw=2)
plot(yy[:,2],rurc,lw=2)
ylim(0,1.2)
semilogx()
```
