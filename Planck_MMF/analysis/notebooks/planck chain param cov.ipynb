```python
params=[Om, Ol, H0, ns, s8, tau, lnA]
```

```python
cov=np.zeros((len(params),len(params)),dtype=np.float64)
for i,p1 in enumerate(params):
    for j,p2 in enumerate(params):
        cov[i,j]=np.mean((d[:,p1]-np.mean(d[:,p1]))*(d[:,p2]-np.mean(d[:,p2])))
```
