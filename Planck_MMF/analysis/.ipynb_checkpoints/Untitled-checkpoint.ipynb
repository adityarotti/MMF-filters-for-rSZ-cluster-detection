```python
from modules.settings import mmf_settings as mmfset
from data_preprocess import preprocess_planck_data as ppd
from flat_sky_codes import tangent_plane_analysis as tpa
import healpy as h
from astropy.io import fits
%pylab notebook
```

```python
mmfset.channels
```

```python
mmfset.init()
```

```python
mask=ppd.gen_ps_mask()
ext_mask=ppd.gen_ps_mask(ps_cutoff=5)
```

```python
mmf3=ppd.get_mmf3_catalogue(snrthr=6,cosmo_flag=True,zknown=True)
for idx,clstrname in enumerate(mmf3["NAME"][:2]):
    glon=mmf3["GLON"][idx]
    glat=mmf3["GLAT"][idx]
    snr=mmf3["SNR"][idx]
    filename=mmfset.paths["tplanes"] + "cluster_" + clstrname[5:] + ".fits"
    projop=tpa.tangent_plane_setup(mmfset.nside,mmfset.xsize,glat,glon,rescale=1.)
    
    timage=projop.get_tangent_plane(mask)
    hdu_mask = fits.ImageHDU()
    hdu_mask.header["EXTNAME"]="Point Source Mask"
    hdu_mask.data=timage
    fits.append(filename,hdu_mask.data,hdu_mask.header)
    
    timage=projop.get_tangent_plane(ext_mask)
    hdu_mask = fits.ImageHDU()
    hdu_mask.header["EXTNAME"]="Extended Point Source Mask"
    hdu_mask.data=timage
    fits.append(filename,hdu_mask.data,hdu_mask.header)
```

```python
f=fits.open(filename)
```
